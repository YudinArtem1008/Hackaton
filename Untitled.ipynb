```python
df_nir = pd.read_csv("train/B8A.csv", sep=',', encoding="windows-1251")
df_swir = pd.read_csv("train/B12.csv", sep=',', encoding="windows-1251")
```

```python
df_nir
```

```python
df_swir = df_swir.drop(columns=["index"])
df_nir = df_nir.drop(columns=["index"])
```

```python
df_ndvi = df_ndvi.dropna()
df_swir = df_swir.dropna()
df_nir = df_nir.dropna()
```

```python
df_features = df_ndvi.copy()

for day in xx:
    df_features[f"{day}_NDWI"] = (df_nir[day] - df_swir[day]) / (df_nir[day] + df_swir[day])
    df_features[f"{day}_NEW"] = df_ndvi[day] * df_features[f"{day}_NDWI"]
```

```python
encoder = OneHotEncoder()

x = df_features.drop(columns=["culture"])
y = pd.DataFrame(encoder.fit_transform(df_features[['culture']]).toarray(), columns=encoder.get_feature_names_out(['culture']))

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train = torch.tensor(np.array(x_train).astype('float32'), dtype=torch.float32)
x_test = torch.tensor(np.array(x_test).astype('float32'), dtype=torch.float32)
y_train = torch.tensor(np.array(y_train).astype('float32'), dtype=torch.float32)
y_test = torch.tensor(np.array(y_test).astype('float32'), dtype=torch.float32)

y_train = torch.tensor(y_train.max(1).indices)
y_test = torch.tensor(y_test.max(1).indices)
y_train.shape, y_test.shape
```

```python
model = torch.nn.Sequential(
    OrderedDict([
        ("linear_1", torch.nn.Linear(78, 200)),
        ("activation_1", torch.nn.ReLU()),
        ("dropout_1", torch.nn.Dropout(p=0.6)),
        ("linear_2", torch.nn.Linear(200, 500)),
        ("activation_2", torch.nn.ReLU()),
        ("dropout_2", torch.nn.Dropout(p=0.6)),
        ("linear_3", torch.nn.Linear(500, 6)),
        ("log_softmax", torch.nn.LogSoftmax(dim=1)),
    ])
)

model
```

```python
epochs = 1000
batch_size = 30
val_every = 10

losses = {"train": [], "val": []}

loss_fn = torch.nn.NLLLoss()

optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.1,
)

for epoch in range(1, epochs + 1):
    optimizer.zero_grad()
    y_pred = model(x_train)
    loss = loss_fn(y_pred, y_train)

    losses["train"].append(loss.item())
    
    loss.backward()
    optimizer.step()
    
    with torch.no_grad():
        if epoch % val_every == 0:
            val_batch_pos = torch.randint(high=x_test.shape[0], size=[batch_size])
            val_pred = model(x_test[val_batch_pos])
            val_loss = loss_fn(val_pred, y_test[val_batch_pos])
            losses["val"].append(val_loss.item())

    if epoch % val_every == 0:
        clear_output(True)
        fig, ax = plt.subplots(figsize=(30, 10))
        plt.title("График ошибки")
        plt.plot(losses["train"], ".-", label="Ошибка на обучении")
        plt.plot(torch.arange(0, epoch, val_every), losses["val"], ".-", label="Ошибка на валидации")
        plt.xlabel("Итерация обучения")
        plt.ylabel("Значение ошибки")
        plt.legend()
        plt.grid()
        plt.show()
```

```python
with torch.no_grad():
    pred_sub = model(x_test).max(1).indices
    acc_score = (y_test == pred_sub).to(dtype=torch.float).mean().item()
    print('Доля правильно классифицированных штук', round(acc_score, 2))
    print(classification_report(y_test, pred_sub))
```
